```python
from dataclasses import dataclass, field  # Import field
import csv
import os
import random
import pandas as pd
import ast
import numpy as np
```

```python
# Load the existing relevant files
ledger = pd.read_csv("Followers List & Categories - Accounts Kept.csv")
og_list = pd.read_csv("followers_following_party_US.csv")

# Select relevant columns
ledger = ledger[["Political Influencer", "Affiliation"]]
og_list = og_list[["Influencer_Xhandle", "Political_Follower_Name", "Political_Followers_XHandle"]]

# Filter the DataFrame for Republican Party affiliation
republican_accounts = ledger[ledger["Affiliation"] == "Republican Party"]["Political Influencer"]
all_accounts = ledger[["Political Influencer"]]

og_list.head()
print(all_accounts)
```

```python
filtered_df = og_list[og_list['Influencer_Xhandle'].isin(all_accounts['Political Influencer'])]

# Get all unique political follower names 
unique_follower_names = filtered_df['Political_Follower_Name'].unique()

# Convert the array to a DataFrame 
# unique_follower_name_df = pd.DataFrame(unique_follower_names, columns=['Political_Follower_Name']) 
# Export the DataFrame to a CSV file 
# unique_follower_name_df.to_csv('Unique Follower Name.csv', index=False)

filtered_df.head()
```

```python
follower_affiliation = pd.read_csv("Unique Follower Name.csv")

follower_dict = follower_affiliation.to_dict('records')
ledger_dict = ledger.to_dict('records')
```

```python
influencer_affiliation_dict = {}
follower_affiliation_dict = {}

for follower in follower_dict:
    follower_affiliation_dict[follower["Political_Follower_Name"]] = follower["Affiliation"]

for influencer in ledger_dict:
    influencer_affiliation_dict[influencer["Political Influencer"]] = influencer["Affiliation"]

filtered_df["Follower Affiliation"] = filtered_df["Political_Follower_Name"].map(follower_affiliation_dict)
filtered_df["Influencer Affiliation"] = filtered_df["Influencer_Xhandle"].map(influencer_affiliation_dict)

filtered_df = filtered_df[["Influencer_Xhandle", "Influencer Affiliation", "Political_Follower_Name", "Political_Followers_XHandle", "Follower Affiliation"]]

# filtered_df.to_csv("Extended Influencers.csv")

kept_filtered_df = filtered_df[filtered_df['Influencer_Xhandle'].isin(ledger['Political Influencer'])]
kept_filtered_df.to_csv("Kept Influencer Extended.csv")
kept_filtered_df.head()
```
