```python
import requests
import pandas as pd
import csv
from datetime import datetime, timedelta
import time

#day by day
date = '20231007'
#increase time
# date = (datetime.strptime(date, '%Y%m%d') + timedelta(days=1)).strftime('%Y%m%d')

# NYT API endpoint and common parameters
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json"

status = True
all_articles = []

page = 0 
while status:
# for page in range(3):  # Retrieve 3 pages (30 articles)
    if date == '20231101':
        break
    try:
        common_params = {
        'q': 'Israel-Gaza War (2023-)',
        'news_desk.contains': 'World OR Politics',
        'begin_date': (date),  # Start date (YYYYMMDD)
        'end_date': (date),  # End date (YYYYMMDD)
        # AND (Trump OR Kamala OR Binkley OR Burgum OR Christie OR Desantis OR Elder OR Hurd OR Hutchinson OR Johnson OR Pence OR Ramaswamy OR Scott OR Suarez OR Haley OR Biden OR Phillips OR Palmer OR Williamson)
        
        'type_of_material': 'News',
        'api-key': 'RvLloVIecCR4YHO3qYT7vg0ggVv3mOAb'
        }

        print(f'Initial date value: {date}')
        print(f'Current page value: {page}')

        params = {**common_params, 'page': page}
        print(f'Common params: {common_params}')

        response = requests.get(url, params=params)
        data = response.json()
        all_articles.extend(data['response']['docs'])
        time.sleep(5)
        page+=1

    except: 
        end_date = (datetime.strptime(date, '%Y%m%d') + timedelta(days=1)).strftime('%Y%m%d')
        date = end_date
        page = 0 
        print(f'Working on {date}')
        continue
        

csv_filename = f'Political Influencer_NYT_{date}_{end_date}'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Headline', 'Publication Date', 'People Mentioned', 'URL'])

    # Display the retrieved articles with mentioned people and organizations
    for article in all_articles:
        article['pub_date'] = pd.to_datetime(article['pub_date'])
        print("Headline:", article['headline']['main'])
        print("Publication Date:", article['pub_date'])
        print("URL:", article['web_url'])
        
        # Extract and display people and organizations mentioned
        persons = [kw['value'] for kw in article['keywords'] if kw['name'] == 'persons']
        organizations = [kw['value'] for kw in article['keywords'] if kw['name'] == 'organizations']
        
        print("People Mentioned:", persons)
        print("Organizations Mentioned:", organizations)
        print()

        writer.writerow([article['headline']['main'], article['pub_date'], persons, article['web_url']])
    
# print(all_articles)
```
