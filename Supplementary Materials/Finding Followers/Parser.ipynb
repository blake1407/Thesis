```python
from dataclasses import dataclass
import csv

new_influencers = []
account_only = []
discarded_accounts = []
accepted_accounts = []

@dataclass
class Influencer:
    account: str
    count: int

#Read in a list of already discarded influencers
discarded_path = "Account Discarded.txt"
with open(discarded_path, 'r') as f:
    for line in f:
        discarded_accounts.append(line.strip().lower())

# #Read a list of already accepted influencers
# accepted_path = "/Users/blakey/Desktop/Reddit-JSON-Parser/Supplementary Materials/Accounts Kept"
# with open(accepted_path, 'r') as f:
#     for line in f:
#         accepted_accounts.append(line.strip().lower())

existing_influencers = []

with open("../Extended Influencers.csv", 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for line in reader:
        account = line[0]
        if account not in existing_influencers:
            existing_influencers.append(line[0].strip().lower())

#Creating a list of newly gathered influencers with their counts of occurence
with open('Followers Parsed.txt', 'r') as f: 
    for line in f:
        if "@" in line:
            line = line.strip().lower()
            # if line not in account_only and line not in discarded_accounts and line not in accepted_accounts:
            if line not in account_only and line not in existing_influencers and line not in discarded_accounts:
                account_only.append(line)
                new_influencers.append(Influencer(line.strip(), 0))
            if line in account_only:
                for i in new_influencers:
                    if i.account == line.strip():
                        i.count +=1

for i in new_influencers:
    if i.count >= 3:
        print(i.account)
        
```
